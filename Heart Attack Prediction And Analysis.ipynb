## 1.1 Examining the project topic <a id = 2 ></a>

### What is a Heart attack?



* A myocardial infarction (commonly called a heart attack) is an extremely dangerous condition caused by a lack of blood flow to your heart muscle.
* A heart attack (myocardial infarction) happens when one or more areas of the heart muscle don't get enough oxygen. This happens when blood flow to the heart muscle is blocked. 

### How Does a Heart Attack Occur?



* A heart attack occurs when an artery that sends blood and oxygen to the heart is blocked

#### What are the symptoms of Heart attack?



## 1.2 Variables in the dataset <a id = 3 ></a>

1. age - age in years


2. sex - sex (1 = male; 0 = female)


3. cp - chest pain type (1 = typical angina; 2 = atypical angina; 3 = non-anginal pain; 0 = asymptomatic)


4. trtbps - resting blood pressure (in mm Hg on admission to the hospital)


5. chol - serum cholestoral in mg/dl


6. fbs - fasting blood sugar > 120 mg/dl (1 = true; 0 = false)


7. restecg - resting electrocardiographic results (1 = normal; 2 = having ST-T wave abnormality; 0 = hypertrophy)


8. thalach - maximum heart rate achieved


9. exang - exercise induced angina (1 = yes; 0 = no)


10. oldpeak - ST depression induced by exercise relative to rest


11. slope - the slope of the peak exercise ST segment (2 = upsloping; 1 = flat; 0 = downsloping)


12. ca - number of major vessels (0-3) colored by flourosopy


13. thal - 2 = normal; 1 = fixed defect; 3 = reversable defect


14. num - the predicted attribute - diagnosis of heart disease (angiographic disease status)(Value 0 = < diameter narrowing; 
Value 1 = > 50% diameter narrowing)

## 5.4 Dealing with outliers <a id = 22 ></a>

```python
IQR = df['trtbps'].quantile(0.75) - df['trtbps'].quantile(0.25)
IQR
```

```python
df['trtbps'].describe()
```

```python
lower_bridge = df['trtbps'].quantile(0.25) - (IQR*1.5)
```

```python
upper_bridge = df['trtbps'].quantile(0.75) + (IQR*1.5)
```

```python
lower_bridge, upper_bridge
```

```python
df.loc[df['trtbps'] < 90, 'trtbps'] = 90
```

```python
df.loc[df['trtbps'] > 170, 'trtbps'] = 170
```

```python
sns.boxplot(df['trtbps'])
```

```python
#thalachh outlier removal

IQR_thalachh = df['thalachh'].quantile(0.75) - df['thalachh'].quantile(0.25)
IQR_thalachh
```

```python
lower_bridge_thalachh = df['thalachh'].quantile(0.25) - (IQR*1.5)

upper_bridge_thalachh = df['thalachh'].quantile(0.75) + (IQR*1.5)
```

```python
lower_bridge_thalachh, upper_bridge_thalachh
```

```python
df.loc[df['thalachh'] < 103.5, 'thalachh'] = 103.5

df.loc[df['thalachh'] > 196.0, 'thalachh'] = 196.0
```

## 5.9 Train Test Split <a id = 25 ></a> 

```python
from sklearn.model_selection import train_test_split

X = df_copy.drop(['output'], axis =1)
y = df_copy['output']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 3)
```

```python
X_test.shape
```

## 5.10 Modelling For Machine Learning <a id = 26 ></a>

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
log_reg = LogisticRegression()
```

```python
log_reg.fit(X_train, y_train)
```

```python
y_pred = log_reg.predict(X_test)
y_pred
```

## 5.11 Accuracy Score <a id = 27 ></a>

```python
print(accuracy_score(y_test, y_pred))
```

## 5.12 Cross Validation Score <a id = 28 ></a>

```python
from sklearn.model_selection import cross_val_score

scores = cross_val_score(log_reg, X_test, y_test, cv = 15)

print(scores.mean())
```
